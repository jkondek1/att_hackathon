```python
df_cumulative = pd.DataFrame(df["Team"].value_counts().cumsum()/df["Team"].shape[0]).reset_index()
counts = df["Team"].value_counts()
```

```python
df_cumulative['absolute'] = list(counts)
```

```python
df_cumulative['percentile'] = df_cumulative['Team']
del df_cumulative['Team']
df_cumulative['Team'] = df_cumulative['index']
del df_cumulative['index']
```

```python
df3 = pd.merge(df, df_cumulative, on = 'Team')
df3.to_csv("version_2.csv", index = False)
```
